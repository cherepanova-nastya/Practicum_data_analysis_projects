## Вывод

- Исследовали выбросы в столбцах Power, Kilometer, Price. Убрали лишние строки.
- Считаю, что столбцы DateCrawled и LastSeen надо удалить, т.к. они не влияют на цену машины. 
- Уберем также DateCreated, т.к. модель не работает с форматом даты.
- Есть пропуски в столбцах: VehicleType, Gearbox, Model, FuelType, NotRepaired. В категориальных признаках есть пропуски, записанные как новые значения: например, None в столбце Gender. Для LightGBM это не проблема, для библиотеки
- Удалим столбец NumberOfPictures, т.к. все значения в нем - 0.
- Изменим тип данных у столбца Price, т.к. для расчета качества модели нам необходимо вычислять квадратный корень - получим дроби.
- Заменим тип данных у категориальных признаков на тип category.
- Заменим пропуски на 'empty', т.к. CatBoost не работает с Nan. 
- Признак PostalCode категориальный, удалим его.
- Добавим столбец с разностью дат, проработаем столбец с месяцем.

```python
#Изменим тип данных у столбцов DateCreated и DateCrawled.
data.loc[:, 'DateCreated'] = pd.DatetimeIndex(data['DateCreated']).year
#Добавим новый столбец с их разностью.
data.loc[:, 'Odometer'] = data['DateCreated'] - data['RegistrationYear']
```

```python
#В данных есть 0-ые месяцы, удалим их.
data.drop(data[data['RegistrationMonth'] ==0].index, inplace=True)
```

```python
#Преобразуем столбец RegistrationMonth.
data['mnth_sin'] = np.sin((data['RegistrationMonth']-1)*(2.*np.pi/12))
data['mnth_cos'] = np.cos((data['RegistrationMonth']-1)*(2.*np.pi/12))

data.drop(['RegistrationMonth'], axis='columns', inplace=True)
```

```python
#Заменим пропуски на 'empty'.
data=data.fillna('empty')
```

```python
#Изменим тип данных у столбца Price.
data['Price'] = data['Price'].astype('float')

#Заменим тип данных у категориальных признаков на тип category.
data['VehicleType'] = data['VehicleType'].astype('category')
data['Gearbox'] = data['Gearbox'].astype('category')
data['Model'] = data['Model'].astype('category')
data['FuelType'] = data['FuelType'].astype('category')
data['Brand'] = data['Brand'].astype('category')
data['NotRepaired'] = data['NotRepaired'].astype('category')
```

```python
#Удалим столбцы DateCrawled, DateCreated и LastSeen.
#Удалим столбец NumberOfPictures, т.к. все значения в нем - 0.
data.drop(['DateCrawled', 'LastSeen', 'DateCreated', 'NumberOfPictures'], axis='columns', inplace=True)
```

```python
#Смотрим общую информацию.
print(data.info())
```

```python
#Удаляю признак PostalCode.
data.drop(['PostalCode'], axis='columns', inplace=True)
```

```python
#Смотрим общую информацию.
print(data.info())
```

# 2. Обучение моделей

## Разобьем данные на выборки

Разобъем данные на 2 выборки: обучающую и тестовую в соотношении: 80/20.

```python
data_train, data_test = train_test_split(data, test_size=0.2, random_state=12345)
```

Для анализа зафиксируем признаки в переменных features_, а целевой признак в target_ для каждой из 2-ух выборок.

```python
#Обучающая выборка.
features_train = data_train.drop(['Price'], axis=1)
target_train = data_train['Price']

#Тестовая выборка.
features_test = data_test.drop(['Price'], axis=1)
target_test = data_test['Price']
```

```python
#Укажем массив категориальных признаков.
cate_features_name = ['VehicleType', 'Gearbox', 'Model', 'FuelType', 'Brand', 'NotRepaired']
cat_features_index = [0, 2, 4, 6, 7, 8]
```

**Обучим модель градиентного бустинга LightGBM.**

```python
#Зададим модель, подобрав параметры с помощью grid_search.
model_lgb = LGBMRegressor(cat_features=cate_features_name) 
parameters = {'max_depth': [5, 7, 9], 'learning_rate': [0.1], 
              'num_leaves': [900], 'n_estimators': [200, 250, 300, 350]}
gs_lgb = GridSearchCV(model_lgb, parameters, n_jobs=-1, scoring='neg_mean_squared_error', cv=3)
gs_lgb.fit(features_train, target_train)
```

```python
#Вывожу качество и лучшие параметры
print('RMSE:', sqrt(-1*gs_lgb.best_score_))
print('Лучшие параметры модели:', gs_lgb.best_estimator_)
```

```python
#Переобучаем модель, использовав подобранные парметры.
start_time = time.time()
model_lgb_1 = LGBMRegressor(cat_features=['VehicleType', 'Gearbox', 'Model', 'FuelType', 'Brand', 'NotRepaired'],
                            learning_rate=0.1, max_depth=9, n_estimators=350, n_jobs=-1, num_leaves=900) 
model_lgb_1.fit(features_train, target_train)

cut_time = time.time()
```

```python
print("Затраченное время (сек) =", cut_time - start_time)
```

**Обучим модель градиентного бустинга CatBoost.**

```python
#Зададим модель, подобрав параметры с помощью grid_search.
model_cbr = CatBoostRegressor(cat_features=cate_features_name, iterations=100) 
grid = {'learning_rate': [0.03, 0.1],
        'depth': [4, 6, 10],
        'l2_leaf_reg': [1, 3, 5, 7, 9]}

grid_search_result = model_cbr.grid_search(grid, cv=3, X=features_train, y=target_train, verbose=False)
```

```python
#Зафиксируем подобранные параметры.
parametrs = grid_search_result['params']

print('Лучшие параметры модели:', parametrs)
```

```python
#Переобучаем модель, использовав подобранные парметры.
start_time = time.time()
model_cbr_1 = CatBoostRegressor(depth = 10, l2_leaf_reg = 1, learning_rate = 0.1, 
                                cat_features=cate_features_name, iterations=100)
model_cbr_1.fit(features_train, target_train, verbose=False)

cut_time = time.time()
```

```python
print("Затраченное время (сек) =", cut_time - start_time)
```

# 3. Анализ моделей

Проверим модель градиентного бустинга CatBoost на тестовой выборке.

```python
#Рассчитаем RMSE для модели.
start_time = time.time()
predictions_test_cbr = model_cbr_1.predict(features_test)
rms_cbr = sqrt(mean_squared_error(target_test, predictions_test_cbr))
cut_time = time.time()

print(rms_cbr)
print("Затраченное время (сек) =", cut_time - start_time)
```

Проверим модель градиентного бустинга LightGBM на тестовой выборке.

```python
start_time = time.time()
predictions_test_lgb = model_lgb_1.predict(features_test)
rms_lgb = sqrt(mean_squared_error(target_test, predictions_test_lgb))
cut_time = time.time()

print(rms_lgb)
print("Затраченное время (сек) =", cut_time - start_time)
```
