## Сборный проект: планирование кампании интернет-магазина «Стримчик»

Необходимо проанализировать данные интернет-магазина «Стримчик» за 2016 год. Также из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы.

**Цель исследования:** выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

#### Описание данных
- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы долларов)
- EU_sales — продажи в Европе (миллионы долларов)
- JP_sales — продажи в Японии (миллионы долларов)
- Other_sales — продажи в других странах (миллионы долларов)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

*Замечания:*
- Данные за 2016 год могут быть неполными.
- В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

Используем функцию для заполнения таких пропусков, которая будет аналогична той, что использовали для year_of_release. Но будем брать медианное значение оценок, так как для одной игры на разных платформах оценка может различаться. Это мы увидели, когда делали срез для игры Inversion.

```python
#Напишем функцию для заполнения таких пропусков.
def fill_critic_na(row):   
    a = data.loc[(data['name'] == row['name']) & (~data['critic_score'].isnull()), 'critic_score'].values
    if(a):
        row['critic_score'] = int(np.median(a))
    return row
```

```python
#Применим функцию.
data.loc[data['critic_score'].isna()] = data.loc[data['critic_score'].isna()].apply(fill_critic_na, axis = 1)
```

Значение tbd обозначает ожидание информации. Соответственно, если осталось это значение, то нет данных о рейтинге. Так как игры с таким значением разных лет, то можно сделать предположение, что этот пропуск и остальные пропуски, рассмотренные до, связанные с рейтингами, существуют потому, что источник сбора данных выдает значение при наборе определенного количества оценок. Если их нет, либо меньше какого-то значения, то проставляется пропуск, либо tbd (зависит от объема).

```python
#Посмотрим на пропуски, ранее мы их заменили на -1.
data.query('user_score == -1')
```

```python
#Посмотрим на строки со значением tbd, ранее мы их заменили на -2.
data.query('user_score == -2')
```

Используем функцию для заполнения таких пропусков, которая будет аналогична той, что использовали для предыдущего столбца. Будем также брать медианное значение оценок, так как для одной игры на разных платформах оценка может различаться. Это мы увидели, когда делали срез для игры Inversion.

```python
#Напишем функцию для заполнения таких пропусков.
def fill_user_na(row):   
    a = data.loc[(data['name'] == row['name']) & (~data['user_score'] == -1) & (~data['user_score'] == -2), 'user_score'].values
    if(a):
        row['user_score'] = int(np.median(a))
    return row
```

```python
#Убираем предупреждения.  
if not sys.warnoptions:
    warnings.simplefilter("ignore")
#Применим функцию.
data.loc[data['user_score'] == -1] = data.loc[data['user_score'] == -1].apply(fill_user_na, axis = 1)
data.loc[data['user_score'] == -2] = data.loc[data['user_score'] == -2].apply(fill_user_na, axis = 1)
```

Выбираем максимальными по прибыли платформы с суммарными продажами более 730 миллионов долларов.

```python
#Составим список максимальных по прибыли платформ.
platform_name = data_platform.index[0:6].values

#Найдем границы лет.
data.query('platform in @platform_name').agg({'year_of_release': ['min', 'max']})
```

```python
#Составим таблицу с данными для топовых платформ.
data_top_pl = pd.DataFrame(index = np.sort(data.query('platform in @platform_name')['year_of_release'].unique().dropna()))

for i in platform_name:
    data_top_pl[i] = data.loc[data['platform'] == i].groupby('year_of_release')['sales'].sum()
```

```python
#Пишем функцию для подсчетов продаж по 3 признакам, которые необходимы по заданию.
#Для дисперсии добавлен параметр ddof=1, т.к. у нас не генеральная совокупность.
def sales_analys(series):
    data_regions = pd.DataFrame()
    for i in series.index:
        row = {}
        row[series.index.name] = i
        row['mean'] = data.loc[data[series.index.name] == i][series.name].mean()
        row['var'] = np.var(data.loc[data[series.index.name] == i][series.name], ddof=1)
        row['deviation'] = np.sqrt(row['var'])
        row['left_b'] = row['mean'] - (3 * row['deviation'])
        if row['left_b'] < 0:
            row['left_b'] = 0
        row['right_b'] = row['mean'] + (3 * row['deviation'])
        data_regions = data_regions.append(row, ignore_index=True)
    data_regions = data_regions[[series.index.name, 'mean', 'var', 'deviation', 'left_b', 'right_b']]
    print('Dataframe for', series.name, ':')
    display(data_regions.sort_values('mean', ascending=False))
    #Строим гистограмму:
    plt.figure(figsize=(12,5), dpi= 80)
    plt.bar(series.index.tolist(), series.values, color=c, width=.5)
    for p, val in enumerate(series.values):
        plt.text(p, val, round(val, 2), horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':12})    
    plt.title(series.name, fontsize=12)
    plt.ylabel('Продажи (млн.$ в год)')
    plt.xlabel(series.index.name)
```

### Шаг 5. Проверка гипотез

Проверим гипотезы и сделаем выводы:
- средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

<font color='purple'>Проверяем гипотезу: "Средние пользовательские рейтинги платформ Xbox One и PC одинаковые"

Формулирую гипотезы для анализа:
- за нулевую принимаю гипотезу: средние пользовательские рейтинги платформ Xbox One и PC одинаковые,
- за альтернативную принимаю гипотезу: средние пользовательские рейтинги платформ Xbox One и PC различаются.

Нулевая гипотеза - это гипотеза равенства, поэтому я так сформировала гипотезы выше.

Для анализа также примем alpha за 5%.

```python
#Формируем выборки для анализа.

data_next = data.drop(data[data['user_score'] < 0].index)

data_user_xo = data_next.query('platform == "XOne"')['user_score']
data_user_pc = data_next.query('platform == "PC"')['user_score']
```

Для выбора критерия ответим на вопрос: близко ли распределение к нормальному?

Для этого:
- строим гистограмму,
- применим метод scipy.stats.normaltest.

```python
#Построим гистограммы средних рейтингов.
fig, ax = plt.subplots(figsize=(12,8), dpi= 80)   
plt.hist(data_user_pc, bins=50, range = (0, 100), label='PC')
plt.hist(data_user_xo, bins=50, range = (0, 100), label='Xbox One')

plt.legend(loc='upper right')
plt.title('Cредние пользовательские рейтинги платформ Xbox One и PC', fontsize=12)
plt.ylabel('Количество')
plt.xlabel('Рейтинги')
plt.show()
```

<h2> Комментарий наставника</h2>

Не забывай подписывать оси на графиках, а также делать им названия. Так твоя работа будет профессиональнее. 

</div>

<font color=blue>**Ответ:** Добавила.</font>

Формулирую гипотезы для анализа:
- за нулевую принимаю гипотезу: распределение близко к нормальному,
- за альтернативную принимаю гипотезу: распределение отличается от нормального.

Проверим это к data_user_xo и data_user_pc.

```python
#Проверяем гипотезу для data_user_xo.
p = st.normaltest(data_user_xo)
alpha = 0.05
print('p-значение:', p.pvalue)
if p.pvalue < alpha:  
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

<div class="alert alert-warning">
<h2> Комментарий наставника 2</h2>

Импорт необходимых библиотек следует произволить в первой ячейке работы, как то рекомендует стандарт. Повторный импорт делать не стоит, он все равно не исполняется, т.к. библиотека уже импортирована в проект. 

</div>

<font color=blue>**Ответ:** убрала.</font>

```python
#Проверяем гипотезу для data_user_pc.
p = st.normaltest(data_user_pc)
alpha = 0.05
print('p-значение:', p.pvalue)
if p.pvalue < alpha:  
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

##### Вывод

Данные противоречат нулевой гипотезе для обеих выборок при заданном уровне значимости alpha. Гипотеза отвергается. Подтвреждение этому видим на гистограмме.

Поэтому применять метод scipy.stats.ttest_ind нельзя. Необходимо применить scipy.stats.mannwhitneyu.

<div class="alert alert-success">
<h2> Комментарий наставника</h2>

Круто, что ты знаешь о таком стат. тесте и умеешь его использовать. 

</div>

<font color='purple'>Чтобы проверить нулевую гипотезу, применим метод scipy.stats.mannwhitneyu.

```python
#Применю метод scipy.stats.mannwhitneyu для проверки гипотезы.
alpha = .05 # критический уровень статистической значимости
            # если p-value окажется меньше него - отвергнем гипотезу

results = st.mannwhitneyu(
    data_user_xo, 
    data_user_pc)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

#### Вывод
Данные не противоречат нулевой гипотезе, не получилось ее отвергнуть. Значит изначальная гипотеза "Средние пользовательские рейтинги платформ Xbox One и PC одинаковые" скорее всего неверна для генеральной совокупности.

Нулевая гипотеза - это гипотеза равенства, поэтому я так сформировала гипотезы выше.

Для анализа также примем alpha за 5%.

Для выбора критерия ответим на вопрос: близко ли распределение к нормальному?

Для этого:
- строим гистограмму,
- применим метод scipy.stats.normaltest.

```python
#Построим гистограммы средних рейтингов.
fig, ax = plt.subplots(figsize=(12,8), dpi= 80)   
plt.hist(data_rt_a, bins=50, range = (0, 100), label='Action')
plt.hist(data_rt_s, bins=50, range = (0, 100), label='Sports')

plt.legend(loc='upper right')
plt.title('Cредние пользовательские рейтинги жанров Action и Sports', fontsize=12)
plt.ylabel('Количество')
plt.xlabel('Рейтинги')
plt.show()
```

Формулирую гипотезы для анализа:
- за нулевую принимаю гипотезу: распределение близко к нормальному,
- за альтернативную принимаю гипотезу: распределение отличается от нормального.

Проверим это к data_rt_a и data_rt_s.

```python
#Проверяем гипотезу для data_rt_a.
from scipy import stats
p = stats.normaltest(data_rt_a)
alpha = 0.05
print('p-значение:', p.pvalue)
if p.pvalue < alpha:  
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

```python
#Проверяем гипотезу для data_rt_s.
from scipy import stats
p = stats.normaltest(data_rt_s)
alpha = 0.05
print('p-значение:', p.pvalue)
if p.pvalue < alpha:  
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```
